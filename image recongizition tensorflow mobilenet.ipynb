```python
import tensorflow as tf
import numpy as np
```

```python
filename='Pictures/zoro.jpg'
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet() ## deep learning mofel weights - pre trained
```

```python
mobile = tf.keras.applications.mobilenet_v2.MobileNetV2() 
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(filename , target_size = (224,224))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
resized_img= image.img_to_array(img)
final_image= np.expand_dims(resized_img,axis =0)
final_image = tf.keras.applications.mobilenet.preprocess_input(final_image)
```

```python
final_image.shape
```

```python
predictions = mobile.predict(final_image)
```

```python
print(predictions)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
results=imagenet_utils.decode_predictions(predictions)
```

```python
print(results)
```
